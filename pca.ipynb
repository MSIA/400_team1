```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA


df = pd.read_csv("new_data.csv")
df = df.drop(columns=['bin', 'avg'])
# scaler = StandardScaler()
# scaler.fit()
```

```python
df.shape
```

```python
df.head(10)
```

```python
scaler = MinMaxScaler()
scaler.fit(df.loc[:, df.columns != 'sku'])
df_sd = scaler.transform(df.loc[:, df.columns != 'sku'])
```

```python
pca = PCA(0.95)
pca.fit(df_sd)
```

```python
df_pca = pd.DataFrame(pca.transform(df_sd))
```

```python
df_pca.head(10)
```

```python
# Plot the explained variances
import matplotlib.pyplot as plt
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_ratio_, color='black')
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)
plt.title("PCA Features by Variance Explained")
```

```python
from sklearn.cluster import KMeans

ks = range(1, 10)
inertias = []
for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k)
    
    # Fit model to samples
    model.fit(df_pca.iloc[:,:3])
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
plt.plot(ks, inertias, '-o', color='black')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.xticks(ks)
plt.title("Clusters vs Inertia")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

hierarchical_cluster = AgglomerativeClustering(affinity='euclidean', linkage='ward')
labels = hierarchical_cluster.fit_predict(df_pca)
labels
```
