```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA


df = pd.read_csv("newBAF_standardized.csv")
df = df.drop(columns = 'Unnamed: 0')
```

```python
df.head(10)
```

```python
import numpy as np
from sklearn.mixture import GaussianMixture
GMM = GaussianMixture(n_components = 7, random_state=0).fit(df.loc[:, df.columns != 'sku']) #7 chosen by pca and kmeans
```

```python
p = GMM.predict(df.loc[:, df.columns != 'sku'])
df["labels"] = p
```

```python
df.head(10)
```

```python
sku_label_df = df[["sku", "labels"]]
sku_label_df.head(10)
```

```python
sku_label_df.to_csv('gmm_labels.csv', index = False)
```
