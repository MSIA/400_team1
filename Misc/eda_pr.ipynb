```python
import pandas as pd
import matplotlib.pyplot as plt

dat = pd.read_csv('purchase_return.csv')
dat.head(10)
```

```python
dat["bin"] = pd.cut(dat["Percentage Return"], 9).astype(str)
dat2 = dat['Percentage Return'].groupby(dat['bin']).count()
p = dat2.plot(kind = 'bar', title='Histogram of Percentage Return')
p.set_ylabel("Frequency")
```
